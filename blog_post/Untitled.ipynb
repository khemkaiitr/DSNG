Question 1: Is there effect of time on bike sharing

```python
# We need to get hour, months, days from the data to analyse effect of timing on bike-sharing data
bike_df.timestamp = pd.to_datetime(bike_df.timestamp)
bike_df['hour'] = bike_df.timestamp.dt.hour
bike_df['day'] = bike_df.timestamp.dt.day
bike_df['month'] = bike_df.timestamp.dt.month
bike_df['year'] = bike_df.timestamp.dt.year
```
